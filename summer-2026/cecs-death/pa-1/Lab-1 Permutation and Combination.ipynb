```python
# Part 1: Permutation (Let's hand-coding first, then use Python itertool)
def generate_permutations(elements):
    results = [] 

    def backtrack(current_path, remaining_options):
        # 1. Base Case: No more options left to pick up
        if len(remaining_options) == 0:
            results.append(current_path)
            return
        # 2. Recursive Step: Iterate through all options
        for i in range(len(remaining_options)):
            # 1) Pick the item
            item = remaining_options[i]
            # 2) Form the new path, i.e., add the item to the current path
            new_path = current_path + [item]
            # 3) For the new option (remove item from remaining)
            new_options = remaining_options[:i] + remaining_options[i+1:] # slicing
            # 4) Recursive call
            backtrack(new_path, new_options)

    # Initial call: path is empty [], options are all elements
    backtrack([], elements)
    return results
```

```python
# Test the backtracking code
input_set = ['A', 'B', 'C', 'D']
perms = generate_permutations(input_set)
print(len(perms))      # Output: 6
print(perms)           # Output: [['A', 'B', 'C'], ['A', 'C', 'B'], ...]
```

```python
# Part 2: Combination (N elements choose k)
def generate_combinations(elements, k): 
    results = []
    def backtrack(start_index, current_path):
        # 1. Base Case: combination is full
        if len(current_path) == k:
            results.append(current_path[:]) # Important: Copy the list
            return 
        # 2. Recursive Step: we only iterate from the start_index to the end
        for i in range(start_index, len(elements)):
            item = elements[i]
            # 3. Action: Choose
            current_path.append(item)
            # 4. Recurse: Move start_index to i+1 (look forward only)
            backtrack(i+1, current_path)
            # 5. Undo the choice
            current_path.pop() 
            
    backtrack(0, [])
    return results
```

```python
# Demo 
data = ['A', 'B', 'C']
k = 2
combos = generate_combinations(data, k)
for c in combos:
    print(c)
```

```python
# Create the iterator
combo_iter = itertools.combinations(data, k)
# Remember: Convert the iterator to a list
combo_list = list(combo_iter)
print(combo_list)
```
